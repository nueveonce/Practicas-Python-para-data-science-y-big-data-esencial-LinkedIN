# KNN 

```python
import pandas as pd
import numpy as np
from sklearn.cluster import KMeans
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv("D:/Datasets/base_datos_2008.csv", nrows=1e5)
# newdf = df[["AirTime","DepDelay"]].dropna()
newdf = df[["AirTime","Distance","TaxiOut","ArrDelay","DepDelay"]].dropna()
```

```python
kmeans = KMeans(n_clusters=3,random_state=0).fit(newdf)
kmeans.labels_
```

```python
np.unique(kmeans.labels_,return_counts=True)
```

```python
plt.scatter(newdf["AirTime"],newdf["DepDelay"],c=kmeans.labels_)
plt.show()
```

```python
kmeans.cluster_centers_
```
