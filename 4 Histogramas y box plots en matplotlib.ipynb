## Histograma y box plots en Matplotlib

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv("D:/Datasets/Base_datos_2008.csv")
```

```python
type(df)
```

```python

# borramos los valores nulos solo de las variables que vamos a estudiar
df.dropna(inplace=True, subset=["ArrDelay","DepDelay","Distance"])
```

```python
sns.displot(x=df["Distance"], kde=False, bins = 100)
```

```python
sns.kdeplot(df["ArrDelay"])
sns.kdeplot(df["DepDelay"])
plt.xlim([-300,300])
```

```python
df2 = df[df["Origin"].isin(["ATL","HOU","IND"])].sample(frac=1).head(500)
```

```python
sns.boxplot(x="DepDelay", y="Origin", data=df2)
plt.xlim(-20,250)
```
